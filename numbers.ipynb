```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
num = np.arange(20)
```

```python
num.reshape((2,-1))
```

```python
s = pd.Series(num)
```

```python
ax = s.plot(ls='', marker='*', mfc='none', ms=20, figsize=(10, 6))
_ = ax.set_yticks(s)
```

```python
s = np.arange(0, 20, 2)
plt.figure(figsize=(10, 6))

plt.scatter(x=s, y=s, s=s*80, c='orange')
plt.axhline(0, c='r', ls='--', lw=2., alpha=.4)
plt.axvline(0, c='r', ls='--', lw=2., alpha=.4)
plt.ylim((-2, 22))

_ = plt.yticks(np.arange(0, 22, 2))
_ = plt.xticks(s)
```
